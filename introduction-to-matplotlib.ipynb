# Making figures with numpy

* Line plot
* Scatter plot 
* Bar plot
* Histogram
* Subplot

```python
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
x
```

```python
fig, ax = plt.subplots() 
ax.plot(x, x**2)
```

```python
fig, ax = plt.subplots() 
ax.scatter(x, np.exp(x))
```

```python
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x))
```

```python
nut_butter_prices = {
    "Almond Butter" : 10,
    "Peanut Butter" : 8,
    "Cashew Butter" : 12
}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title="Nut Butter Shop", xlabel="nut", ylabel="Price")
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()))
```

```python
x = np.random.randn(999)
fig, ax = plt.subplots()
ax.hist(x)
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(ncols=2, nrows=2, figsize=(10,5))
ax1.plot(x, x/2)                            
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax4.hist(np.random.randn(999))
```

```python
fig, ax = plt.subplots(ncols=2, nrows=2, figsize=(10,5))
ax[0,0].plot(x, x/2)                            
ax[0,1].scatter(np.random.random(10), np.random.random(10))
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1,1].hist(np.random.randn(999))
```

# Plotting from pandas DataFrame

```python
import pandas as pd
```

```python
car_sales = pd.read_csv("car-sales.csv")
car_sales
```

```python
ts = pd.Series(np.random.randn(1000), index=pd.date_range("1/1/2020",periods=1000))
ts = ts.cumsum()
ts.plot()
```

```python
car_sales.Price = car_sales.Price.str.replace(r'[\$\,\.]','', regex=True)
car_sales
```

```python
car_sales.Price = car_sales.Price.str[:-2]
car_sales
```

```python
car_sales["Sales Date"] = pd.date_range("1/1/2020", periods=len(car_sales))
car_sales
```

```python
car_sales["Total Sales"] = car_sales.Price.astype(int).cumsum()
car_sales
```

```python
car_sales.plot(x="Sales Date", y="Total Sales")
```

```python
car_sales.Price = car_sales.Price.astype(int)
car_sales.plot(x="Odometer (KM)", y="Price", kind="scatter")
```
